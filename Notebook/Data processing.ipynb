```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from numpy import genfromtxt
```

```python
thorup_real_world = genfromtxt('thorup_real_world_collected.csv', delimiter=',')
wulff_real_world = genfromtxt('wulff_real_world_collected.csv', delimiter=',')

thorup_complete = genfromtxt('thorup_complete_graph_collected.csv', delimiter=',')
thorup_dense = genfromtxt('thorup_dense_graph_collected.csv', delimiter=',')
thorup_semi_dense = genfromtxt('thorup_semi_dense_graph_collected.csv', delimiter=',')
thorup_sparse = genfromtxt('thorup_sparse_graph_collected.csv', delimiter=',')
thorup_very_sparse = genfromtxt('thorup_very_sparse_graph_collected.csv', delimiter=',')
thorup_super_dense = genfromtxt('thorup_super_dense_graph_collected.csv', delimiter=',')
thorup_very_dense = genfromtxt('thorup_very_dense_graph_collected.csv', delimiter=',')

wulff_complete = genfromtxt('wulff_complete_graph_collected.csv', delimiter=',')
wulff_dense = genfromtxt('wulff_dense_graph_collected.csv', delimiter=',')
wulff_semi_dense = genfromtxt('wulff_semi_dense_graph_collected.csv', delimiter=',')
wulff_sparse = genfromtxt('wulff_sparse_graph_collected.csv', delimiter=',')
wulff_very_sparse = genfromtxt('wulff_very_sparse_graph_collected.csv', delimiter=',')
wulff_super_dense = genfromtxt('wulff_super_dense_graph_collected.csv', delimiter=',')
wulff_very_dense = genfromtxt('wulff_very_dense_graph_collected.csv', delimiter=',')
```

```python
real_world_k_values = [2, 3, 5, 8, 10, 13, 15, 18]
density_k_values = [2, 3, 5, 8, 10, 13]
```

```python
# Thorup real world
t_r_max_qt = []
t_r_min_qt = []
t_r_avg_qt = []
t_r_max_stretch = []
t_r_min_stretch = []
t_r_avg_stretch = []

rd = []

for k in real_world_k_values:
    k_data = thorup_real_world[(k == thorup_real_world[:,2])]
    r, _ = k_data.shape
    qt = k_data[:,1]
    stretch = k_data[:,0]
    
    rd.append(stretch)
    
    t_r_min_qt.append(qt.min())
    t_r_max_qt.append(qt.max())
    t_r_avg_qt.append(qt.mean())
    t_r_max_stretch.append(stretch.max())
    #min_stretch.append(stretch.min)
    t_r_avg_stretch.append(stretch.mean())
    
print(t_r_max_stretch)
print(t_r_avg_stretch)
    
fg1, ax1 = plt.subplots()
fg2, ax2 = plt.subplots()

ax1.set(
    xlabel='K value',
    ylabel='Stretch'
)

ax2.set(
    xlabel='K value',
    ylabel='Stretch'
)

ax1.boxplot(rd, 0, '', positions=real_world_k_values)
ax2.boxplot(rd, positions=real_world_k_values)

plt.show()
```

```python
plt.scatter(real_world_k_values, np.array(t_r_avg_stretch))
plt.xticks(real_world_k_values)
plt.show()
```
