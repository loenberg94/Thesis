```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from numpy import genfromtxt
```

```python
thorup_real_world = genfromtxt('thorup_real_world_collected.csv', delimiter=',')
wulff_real_world = genfromtxt('wulff_real_world_collected.csv', delimiter=',')
thorup_complete = genfromtxt('thorup_complete_collected.csv', delimiter=',')
```

```python
real_world_k_values = [2, 3, 5, 8, 13, 18]
density_k_values = [2, 3, 5, 8, 13]
```

```python
# Thorup real world
max_qt = []
min_qt = []
avg_qt = []
max_stretch = []
min_stretch = []
avg_stretch = []

rd = []

for k in real_world_k_values:
    k_data = thorup_real_world[(k == thorup_real_world[:,3])]
    r, _ = k_data.shape
    expected = k_data[:,0]
    actual = k_data[:,1]
    qt = k_data[:,2]
    stretch = actual / expected
    
    rd.append(stretch)
    
    min_qt.append(qt.min())
    max_qt.append(qt.max())
    avg_qt.append(qt.mean())
    #max_stretch.append(stretch.max)
    #min_stretch.append(stretch.min)
    avg_stretch.append(stretch.mean())
    
fg, axs = plt.subplots(2)    
axs[0].set(
    xlabel='k',
    ylabel='stretch'
)

axs[1].set(
    xlabel='k',
    ylabel='stretch'
)

axs[0].boxplot(rd, 0, '', positions=real_world_k_values)
axs[1].boxplot(rd, positions=real_world_k_values)
#ax.xlabel('k')
#ax.ylabel('stretch')
plt.show()
```
